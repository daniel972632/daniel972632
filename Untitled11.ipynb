```python
# LLM Comparison Chart (Matplotlib / Colab-ready)
# ------------------------------------------------
# Fields (x-axis):    General Use, Long-text Analysis, Search & Info, Coding Support, Productivity
# Services (series):  ChatGPT, Claude, Gemini, Perplexity, xAI (Grok)

import matplotlib.pyplot as plt
import numpy as np

# 1) Data (edit scores 1–10 as you like)
services = ["ChatGPT", "Claude", "Gemini", "Perplexity", "xAI (Grok)"]
fields = ["General Use", "Long-text Analysis", "Search & Info", "Coding Support", "Productivity"]

scores = {
    "ChatGPT":     [9, 8, 7, 9, 9],
    "Claude":      [8, 10, 6, 10, 8],
    "Gemini":      [7, 7, 9, 7, 9],
    "Perplexity":  [6, 6, 10, 6, 7],
    "xAI (Grok)":  [7, 7, 8, 9, 6],
}

# 2) Layout
x = np.arange(len(fields))
num_series = len(services)
bar_width = 0.16
offsets = (np.arange(num_series) - (num_series - 1) / 2) * bar_width

plt.figure(figsize=(12, 6))
ax = plt.gca()

# 3) Bars + labels
for i, svc in enumerate(services):
    y = scores[svc]
    bars = ax.bar(x + offsets[i], y, width=bar_width, label=svc)
    # value labels on top
    for b in bars:
        ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.15,
                f"{b.get_height():.0f}", ha="center", va="bottom", fontsize=9)

# 4) Aesthetics
ax.set_title("Comparison of Conversational AI (LLM) Services", fontsize=14, pad=12)
ax.set_ylabel("Score (1–10)", fontsize=11)
ax.set_xlabel("Field", fontsize=11)
ax.set_xticks(x)
ax.set_xticklabels(fields, rotation=0)

ax.set_ylim(0, 11)
ax.grid(axis="y", linestyle="--", alpha=0.35)
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

# Put legend outside for readability
leg = ax.legend(title="Services", loc="upper left", bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()

# 5) Save high-res PNG (for slides)
plt.savefig("/content/llm_comparison.png", dpi=200, bbox_inches="tight")
plt.show()

print("Saved figure to /content/llm_comparison.png")
```

```python
# LLM Comparison - One chart per field (Matplotlib / Colab-ready)

import matplotlib.pyplot as plt
import numpy as np
import re

# -------------------------
# 1) Data (edit as needed)
# -------------------------
services = ["ChatGPT", "Claude", "Gemini", "Perplexity", "xAI (Grok)"]
fields = ["General Use", "Long-text Analysis", "Search & Info", "Coding Support", "Productivity"]

scores = {
    "ChatGPT":     [9, 8, 7, 9, 9],
    "Claude":      [8, 10, 6, 10, 8],
    "Gemini":      [7, 7, 9, 7, 9],
    "Perplexity":  [6, 6, 10, 6, 7],
    "xAI (Grok)":  [7, 7, 8, 9, 6],
}

# -------------------------
# 2) Helper to slugify filename
# -------------------------
def slugify(text):
    return re.sub(r"[^a-z0-9]+", "_", text.lower()).strip("_")

# -------------------------
# 3) Plot each field separately
# -------------------------
for j, field in enumerate(fields):
    # Collect this field's scores across services
    y = [scores[s][j] for s in services]
    x = np.arange(len(services))
    bar_width = 0.6

    plt.figure(figsize=(8, 5))
    ax = plt.gca()

    bars = ax.bar(x, y, width=bar_width)

    # Value labels on top
    for b in bars:
        ax.text(b.get_x() + b.get_width()/2,
                b.get_height() + 0.15,
                f"{b.get_height():.0f}",
                ha="center", va="bottom", fontsize=10)

    # Titles & labels
    ax.set_title(f"LLM Comparison — {field}", fontsize=14, pad=10)
    ax.set_ylabel("Score (1–10)", fontsize=11)
    ax.set_xticks(x)
    ax.set_xticklabels(services, rotation=0)

    # Style
    ax.set_ylim(0, 11)
    ax.grid(axis="y", linestyle="--", alpha=0.35)
    for spine in ["top", "right"]:
        ax.spines[spine].set_visible(False)

    plt.tight_layout()

    # Save & show
    out_path = f"/content/llm_{slugify(field)}.png"
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"Saved: {out_path}")
```
